Executive Summary:

The goal of this analysis is to determine the significant factors that affect the revenue of Apprentice Chef meal delivery company, and to develop a model that predicts revenue using these factors. After analyzing the given dataset, it was discovered that there are some key factors that affect the revenue of the company. These factors include the total number of meals ordered, the average time per site visit, the customer service contacts, the average mean rating, the total number of photos viewed, and the unique meals purchased. Using these factors, a decision tree regression model was developed with a test score of 71.6%, which is considered to be an accurate model.

Introduction:

Apprentice Chef, Inc. is a company that caters to busy professionals by providing daily gourmet meals that can be ordered through their user-friendly online platform or mobile app. The meals take less than 30 minutes to finish cooking and are priced between $10 and $23. Additionally, the company offers master cooking classes, a weekly meal plan with discounts, and disposable cookware. The platform allows customers to filter meals by cuisine and provides personalized recommendations based on previous purchases, similarity to other customers, and trending products. The company's mission is to support the busy lifestyle of working professionals by offering delicious and healthy eating options. Convenience, variety, and healthy choices are the key value propositions of the company, as determined by market research.
The meal delivery business has become increasingly popular over the years due to its convenience and affordability. The industry is highly competitive and there is a need for meal delivery companies to understand the factors that affect their revenue. In this analysis, we aim to identify these factors and build a model that predicts the revenue of Apprentice Chef meal delivery company.
A comprehensive analysis was carried out to determine the factors affecting the company's revenue. This involved examining the correlation between various variables and revenue. Data showed that some variables had positive correlations while others were negatively correlated or not related. To gain further insights, histogram of each variable was used to assess the skewness of the data. The OLS model from the stats model library was then used to analyze the relationship between revenue and several variables, with factors such as p-value, R-square, adjusted R square, F-statistic and probability of F-statistic and coefficients were monitored, with variables with high p-value taken out of the model. 
Machine Learning Model was developed, after testing various Regression models, and parameters were tuned using the Machine learning hyperparmeter models, to determine the optimal parameter to be used for the model.

Data Analysis:

I began my analysis by examining the given dataset to understand the variables and identify missing values. The dataset consists of 1946 rows and 19 variables, which includes variables related to customer profiles, meal orders, and revenue. I found missing values in the family name column of the dataset. This was replaced with the word ‘Unknown’. 
Categorical variables were not used for this analysis. I then conducted a correlation analysis to determine the variables that have the most significant impact on revenue. The variables that were found to have the strongest correlation with revenue were the total number of meals ordered, the average time per site visit, the customer service contacts, the average mean rating, the total number of photos viewed, and the unique meals purchased. I noticed during my analysis that Average meals ordered had a positive correlation with revenue, but a negative coefficient on the regression model. Further analysis showed multicollinearity effect with average time per site visit. This was the reason the feature was taken out of the model. The variables were corrected for skewness and the logarithmic variables were used for the analysis.

Model Development:

Based on the results of the correlation analysis, I developed a decision tree regression model to predict revenue using these significant factors. The model was developed using scikit-learn, which is a machine learning library in Python. The model was trained using a training dataset consisting of 75% of the data, while the remaining 25% of the data was used for testing.
Our first regression model used the six variables mentioned above to predict revenue. After log-transforming both the response and predictor variables, the final model had an R-squared value of 0.726. This means that the model explains 72.6% of the variation in revenue, which is a relatively strong fit.
Our decision tree regression model had a slightly lower R-squared value of 0.723. This model had an optimal tree structure with a minimum of 26 samples per leaf and a maximum depth of eight. The best splits were determined using the squared error criterion. Despite the lower R-squared value, the decision tree model had a smaller difference between the training and testing scores, indicating that it may generalize better to new data.

The model was tuned using the best parameters to obtain an optimal model. The best parameters were found to be the best split, minimum samples per leaf, maximum depth, and criterion. The model was then fit to the training data and the test set was used to evaluate the model performance.

Results:

The decision tree regression model developed had a test score of 71.6%, indicating that it is a good model for predicting revenue. The model was found to be accurate and consistent, as evidenced by the difference between the test and training scores. The difference was only 0.05, indicating that the model is not overfitting.
Our analysis of the Apprentice Chef dataset found that the top five variables associated with revenue were the average rating, total meals ordered, unique meals purchased, contacts with customer service, and total photos viewed. After developing regression and decision tree models using these variables, we found that both models had strong fits. The regression model had a slightly higher R-squared value, while the decision tree model may generalize better to new data. Overall, these results suggest that the five variables identified in this study could be useful in predicting the revenue generated by customers.
There are several limitations to this analysis. First, the data only includes information on meals ordered by customers, and does not take into account other factors that could affect revenue, such as marketing campaigns or external economic factors. Second, the dataset only includes information on customers in the United States, so the results may not be generalizable to other countries.
Finally, the top five variables identified in this study could be used in future marketing campaigns to target customers who are more likely to generate higher revenue.

Conclusion:

The meal delivery business is highly competitive, and companies need to understand the factors that affect their revenue.
```python
########################################################
#MODEL DEVELOPMENT
#######################################################

from sklearn.model_selection import train_test_split # train/test split
import sklearn.linear_model # linear modeling in scikit-learn
```

```python
# x-variables for Log_y
x_var = ['log_TOTAL_MEALS_ORDERED','log_AVG_TIME_PER_SITE_VISIT','log_CONTACTS_W_CUSTOMER_SERVICE',
         'log_AVG_MEAN_RATING','log_TOTAL_PHOTOS_VIEWED','log_UNIQUE_MEALS_PURCH']
```

```python
# preparing for scikit-learn

# preparing x-variables from the OLS model
x_data = chef.loc[ : , x_var]

# preparing response variable
y_data      = chef.loc[ : , 'log_REVENUE']


###############################################
## setting up more than one train-test split ##
###############################################
x_train, x_test, y_train, y_test = train_test_split(
            x_data, # x-variables (can change this)
            y_data, # y-variable  (can change this)
            test_size    = 0.25,
            random_state = 219)
```

```python
### Testing Model with Linear Regression
# Setting a model name
model_name = "Linear Regression"


# INSTANTIATING a model object - CHANGE THIS AS NEEDED
model = sklearn.linear_model.LinearRegression()


# FITTING to the training data
model_fit = model.fit(x_train, y_train)


# PREDICTING on new data
model_pred = model.predict(x_test)


# SCORING the results
model_train_score = model.score(x_train, y_train).round(4) # using R-square
model_test_score  = model.score(x_test, y_test).round(4)   # using R-square
model_gap         = abs(model_train_score - model_test_score).round(4)


# displaying results
print('Training Score :', model_train_score)
print('Testing Score  :', model_test_score)
print('Train-Test Gap :', model_gap)
```

```python
### Testing Model with ARD Regression
# Setting a model name
model_name = "ARD"


# INSTANTIATING a model object - CHANGE THIS AS NEEDED
model = sklearn.linear_model.ARDRegression()


# FITTING to the training data
model_fit = model.fit(x_train, y_train)


# PREDICTING on new data
model_pred = model.predict(x_test)


# SCORING the results
model_train_score = model.score(x_train, y_train).round(4) # using R-square
model_test_score  = model.score(x_test, y_test).round(4)   # using R-square
model_gap         = abs(model_train_score - model_test_score).round(4)


# displaying results
print('Training Score :', model_train_score)
print('Testing Score  :', model_test_score)
print('Train-Test Gap :', model_gap)
```
