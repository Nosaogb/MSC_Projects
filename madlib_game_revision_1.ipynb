The 3 stage text adventure game based on the movie Prison Break is an exciting and thrilling experience that immerses players into the world of Michael Scofield and his mission to break out of prison and clear his brother's name. In stage 1, players take on the role of Michael and must gather information and resources to help him plan his escape. The player must explore the prison and interact with other inmates to gather clues and items that will aid in the escape plan. Along the way, players will encounter challenges and obstacles that they must overcome to progress. In stage 2, the player puts the escape plan into action, and must navigate through the prison, avoid detection and overcome obstacles to reach the outside. This stage requires the player to use stealth, intelligence and problem-solving skills to escape undetected. The player must also make quick and strategic decisions to avoid getting caught. In stage 3, Michael is on the run from the authorities and must use his cunning and intelligence to evade capture and clear his brother's name. In this stage, the player must solve puzzles and make decisions that will determine Michael's fate. They will encounter different scenarios and challenges that will test their ability to think critically and make decisions on the fly. The game is a combination of problem solving, puzzle-solving and decision-making as the player controls Michael in his quest for freedom. The game is designed to keep the players engaged and on the edge of their seats as they experience the same adrenaline-pumping action and tension of the movie. The game's story is based on the movie and it's a great way to experience the story if you are a fan of the movie.
```python
#importing random function
import random

# Introduction to the game
input(prompt="Welcome to the Prison Break text adventure game. Hit Any key to continue:\n ")


# Defining win function
def win():
    winner = input(prompt=f"What is your name \n").capitalize()
    print(f" Congrats {winner}, You have won the Game!!! ")


# Defining fail function
def fail():
    loser = input(prompt=f"What is your name \n").capitalize()
    print(f" Sorry {loser}, You lost the game")


# Defining Stage 1 and its requirements
def stage_1():
    current_stage = 1
    print(f"Welcome to stage {current_stage}. In this stage, you are required to select a source of information: \n")
    
    information = ["Prison map", "History Book", "Google search", "ChatGpt"]
    for i in information:
        print(i)

    # Creating loop to output 3 tries for wrong answers
    correct_answer = "prison map"
    count = 3

    # Adding random.randint() function to select a random correct answer
    rand_num = random.randint(0, len(information) - 1)
    correct_answer = information[rand_num].lower()

    while count > 0:
        # Controlling Error using Try and Except commands
        try:
            info_1 = input(prompt=f"Which source of information will you require for planning an escape:\n").lower()

            if info_1 == correct_answer:
                print("Excellent choice, You have won this stage!!!")
                break

            # Creating Alternatives and outputs for wrong choice
            else:
                print("Wrong choice, try again.")
                count -= 1

                if count == 0:
                    print("You lose!!!")

        # Controlling Errors(Invalid input) using except syntax
        except ValueError:
            print("Invalid input, please try again.")
            count -= 1

            if count == 0:
                print("You lose!!!")

# Running the function
stage_1()

# Defining Stage 2 and its requirements
def stage_2():
    current_stage = 2
    print(f"Welcome to stage {current_stage}.")
    print("Your goal is to navigate through the prison, avoid detection and overcome obstacles to reach the outside.")
    
    # Defining variables
    current_location = "cell"
    locations = {"yard": "Wrong step, try again", "kitchen": "Wrong step, try again", "escape room": "Right Step, plan for your next move!"}
    reached_escape_room = False
    test_trials = 3
    
    # looping to prevent infinite attempts
    while test_trials > 0:
        # Controlling errors using try and except commands to control for errors
        try:
            actions = input("What would you like to do? (move, hide, overcome obstacle)\n").lower()
            
            if actions == "move":
                destination = input(prompt=f"Where would you like to move to? {', '.join(locations.keys())} \n").lower()
                #Subnesting within a loop, to determine next right move in the game
                if destination in locations:
                    current_location = destination
                    print(locations[destination])
                    if destination == "escape room":
                        reached_escape_room = True
                        #Subnesting within a loop, to determine next right move in the game
                        obstacle = input(prompt="In the escape room, there are obstacles, what will you do? (Overcome obstacle, Move or Hide)\n").lower()
                        if obstacle == "overcome obstacle":
                            # Randomly determine if the player successfully overcomes the obstacle
                            if random.randint(0, 1) == 1:
                                print("Good choice, next step")
                                input(prompt=f"< Hit Enter Key for next step>")
                                print("You have successfully overcome all obstacles and reached the outside. You have won the stage!")
                                break
                            else:
                                print("Oops, you were caught by the guards while trying to overcome the obstacle. You lose!")
                                break
                        #Alternatives for wrong choice (with limited attempts)
                        else:
                            print("Wrong Choice")
                            test_trials -= 1
                    else:
                        print("Wrong Choice")
                        test_trials -= 1
                #Alternatives for wrong choice (with limited attempts)
                else:
                    print("Invalid Option")
                    test_trials -= 1
                    
            elif actions == "hide" or actions == "overcome obstacle":
                if reached_escape_room == True:
                    print("Wrong Choice, Try again")
                    test_trials -= 1
                #Alternatives for invalid choice
                else:
                    print("Invalid Choice, Please move to escape room first")
            else:
                print("Invalid option")
                test_trials -= 1
        #Using Except command to cater for invalid inputs
        except Exception as e:
            print(f'Invalid input. {e}')
            test_trials -= 1
            
    if test_trials == 0:
        print("You lose.!!!")

# running the function
stage_2()

# Defining Stage 2 and its requirements
def stage_3():
    current_stage = 3
    print(f"Welcome to stage {current_stage}.")
    print("You have successfully escaped the prison, but now you must evade capture by the authorities.")
    # Defining variables
    current_location = "outside"
    locations = {"city": "Wrong move, try again", "forest": "Right move, keep going!", "mountains": "Right move, keep going!", "beach": "Right move, keep going!"}
    reached_safe_zone = False
    test_trials = 3
    # looping to prevent infinite attempts
    while test_trials > 0:
        # Try function, for controlling user input error type
        try:
            actions = input(prompt=f"What would you like to do? (move, hide, bribe)\n").lower()
            if actions == "move":
                destination = input(prompt=f"Where would you like to move to? {', '.join(locations.keys())} \n").lower()
                # Subnesting within a loop, to determine the next right move in the stage
                if destination in locations:
                    current_location = destination
                    print(locations[destination])
                    if destination == "mountains" or destination == "beach":
                        reached_safe_zone = True
                        print("You have successfully evaded capture by the authorities and reached a safe zone. You have won the stage!")
                        win()
                        break
                    # Alternatives for wrong choice (with limited attempts)
                    else:
                        print("Wrong Choice")
                        test_trials -= 1
                        # Adding random number of retries
                        test_trials += random.randint(0, 2)
                # Alternatives for invalid option
                else:
                    print("Invalid Option")
                    test_trials -= 1
                    # Adding random number of retries
                    test_trials += random.randint(0, 2)
            # Alternatives for other choices other than move (Choosing hide option) with limited attempts for wrong choices
            elif actions == "hide":
                if reached_safe_zone == True:
                    print("You are already in a safe zone, no need to hide.")
                else:
                    print("You are hiding, but authorities are getting closer...")
                    test_trials -= 1
                    # Adding random number of retries
                    test_trials += random.randint(0, 2)
            # Alternatives for other choices other than move (Choosing bribe option) with limited attempts for wrong choices
            elif actions == "bribe":
                if reached_safe_zone == True:
                    print("You are already in a safe zone, no need to bribe.")
                else:
                    # Controlling user input to ensure it is converted to integer
                    bribe_amount = int(input(prompt=f"How much money would you like to bribe the authorities with?\n"))
                    if bribe_amount > 1000:
                        print("The authorities have been bribed successfully, you can continue to move.")
                    else:
                        print("The bribe amount is too low, the authorities are still chasing you.")
                        test_trials -= 1
                        # Adding random number of retries
                        test_trials += random.randint(0, 2)
            # Alternatives for wrong choice (with limited attempts)
            else:
                print("Invalid option")
                test_trials -= 1
                # Adding random number of retries
                test_trials += random.randint(0, 2)
        # Using the Except command to control user input error
        except Exception as e:
            print(f'Invalid input. {e}')
            test_trials -= 1
    if test_trials == 0:
        print("You lose.!!!")
        fail()
        
# running the function
stage_3()
```
